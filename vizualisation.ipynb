```python
# 1. Authenticate to Azure Storage
storage_account_name = "{storage_account_name}" 
storage_account_key = "{storage_account_key}" 
spark.conf.set(f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net", storage_account_key)

# 2. Define the Gold Path
gold_path = f"abfss://dubairealestatedata@{storage_account_name}.dfs.core.windows.net/gold"

# 3. Register Managed Tables for the BI Layer
# Table 1: Sales
spark.read.format("delta").load(f"{gold_path}/sales_overview").write.mode("overwrite").saveAsTable("gold_sales_overview")
# Table 2: Area Stats
spark.read.format("delta").load(f"{gold_path}/area_stats").write.mode("overwrite").saveAsTable("gold_area_stats")
# Table 3: Supply Stats
spark.read.format("delta").load(f"{gold_path}/supply_stats").write.mode("overwrite").saveAsTable("gold_supply_stats")

print("✅ Power BI Bridge: All 3 Tables Registered Successfully.")
```
